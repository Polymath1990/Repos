```python
def road(speedo,birthday,public_hols):
    if public_hols:
        speed = speedo + 10
    elif birthday:
        speed = speed - 5
    else: speed = speed
        
    
        
    if speed > 100:
        return 'big ticket'
    elif speed > 80:
        return 'small ticket'
    elif speed > 60:
        return 'first offender'
    else:
        return 'no ticket'
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']},
                     index = ['K0', 'K0', 'K1', 'K2'])
    
right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']},
                               index = ['K0', 'K1', 'K2', 'K3'])
```

```python
pd.merge(left,right,how ='inner', on ='key1')
```

```python
left.join(right, how = 'outer')
```

```python
pd.concat([left,right], axis = 1)
```
